Simple Notebook to plot accuracy vs epochs

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# load csv containing the graph
l1 = pd.read_csv("lemma.csv")
l2 = pd.read_csv("words.csv")
l3 = pd.read_csv("noparty.csv")

l1["Step"] = l1["Step"].apply(lambda x: x + 1)
l2["Step"] = l2["Step"].apply(lambda x: x + 1)
l3["Step"] = l3["Step"].apply(lambda x: x + 1)

#plot accuracy per epoch
plt.plot(l2["Step"], l2["Value"], color = "red", label = "Em. Matrix (word token)")
plt.plot(l3["Step"], l3["Value"], color = "green", label = "Em. Matrix (no party)")
plt.plot(l1["Step"], l1["Value"], color = "blue", label = "Em. Matrix (lemma)")
plt.xlabel("Epoch")
plt.ylabel("Accuracy (Categorical Crossentropy, in %)");
plt.legend(loc = 5);
plt.grid(which = "minor")
plt.text(l1["Step"][14], l1["Value"][14]-0.004, "76.98%")
plt.text(l2["Step"][14], l2["Value"][14]-0.004, "82.23%")
plt.text(l3["Step"][14], l3["Value"][14]-0.004, "80.10%")
plt.xlim(0.8, 15)
plt.savefig("matrix_acc.png")
```
